```python
questions=["1. What is the distribution of the perimeter mean of breast cancer tumors?",
           "2. What is the distribution of the smoothness mean of breast cancer tumors?",
           "3. What is the relationship between the mean radius and mean area of breast cancer tumors?",
           "4. What is the relationship between the mean texture and mean smoothness of breast cancer tumors?"]
questions
           
```

```python
# 1.Distribution of Perimeter Mean
plt.figure(figsize=(8, 6))
sns.histplot(df['perimeter_mean'], bins=30, kde=True, color='purple')
plt.title('Distribution of Perimeter Mean')
plt.xlabel('Perimeter Mean')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2.Distribution of Smoothness Mean
plt.figure(figsize=(8, 6))
sns.histplot(df['smoothness_mean'], bins=30, kde=True, color='orange')
plt.title('Distribution of Smoothness Mean')
plt.xlabel('Smoothness Mean')
plt.ylabel('Frequency')
plt.show()
```

```python
# 3.Relationship between Mean Radius and Mean Area
plt.figure(figsize=(8, 6))
sns.scatterplot(x='radius_mean', y='area_mean', hue='diagnosis', data=df, palette='coolwarm')
plt.title('Relationship between Mean Radius and Mean Area')
plt.xlabel('Mean Radius')
plt.ylabel('Mean Area')
plt.show()
```

```python
# 4.Relationship between Mean Texture and Mean Smoothness
plt.figure(figsize=(8, 6))
sns.scatterplot(x='texture_mean', y='smoothness_mean', hue='diagnosis', data=df, palette='coolwarm')
plt.title('Relationship between Mean Texture and Mean Smoothness')
plt.xlabel('Mean Texture')
plt.ylabel('Mean Smoothness')
plt.show()
```
