```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# reading the database
data = pd.read_csv("tips.csv")
```

```python
# print ing the top 10 rows
display (data.head(10)) 
```

```python
# Scatter plot with day against
plt.scatter ( data['day'], data['tip'])
# Adding Tit le to the Plot
plt.title("Scatter Plot" )
# Setting the X and Y labels
plt.xlabel('Day')
plt.ylabel ('Tip')
plt.show( ) 
```

```python
# Scatter plot with day against
plt.scatter(data['day'], data['tip'], s=data['total_bill'],c=data['total_bill'],cmap='viridis')
# Adding Tit le to the Plot
plt.title (" Scatter Plot")
# Setting the X and Y labels
plt.xlabel ( 'Day')
plt.ylabel('Tip' )
plt.colorbar ()
plt.show () 
```

```python
# Scatter plot with day against
plt.plot (data['tip'])
plt.plot (data['size'])
```

```python
# Bar chart with day against ti
plt.bar (data['day'], data ['tip'])
plt.title("Bar Chart")
# Setting the X and Y labels
plt.xlabel ('Day' )
plt.ylabel( 'Tip')
# Adding the legends
plt. show() 
```

```python
# Bar chart with day against ti
plt.hist(data['total_bill'], bins=20,edgecolor='black')
plt.title("Bar Chart")
# Setting the X and Y labels
plt.xlabel ('total bill' )
plt.ylabel( 'frequency')
# Adding the legends
plt. show() 
```
